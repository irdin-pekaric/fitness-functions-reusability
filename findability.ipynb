```python
import json
import requests

class FindabilityChecker:
    def __init__(self, json_file):
        self.metadata_repository = self.load_metadata(json_file)

    def load_metadata(self, json_file):
        with open(json_file, 'r') as file:
            data = json.load(file)
        return {artifact_id: artifact['uri'] for artifact_id, artifact in data['artifacts'].items()}

    def is_online(self, uri):
        try:
            response = requests.head(uri, allow_redirects=True)
            return response.status_code == 200
        except requests.RequestException:
            return False

    def check_findability(self):
        for artifact_id, uri in self.metadata_repository.items():
            if self.is_online(uri):
                print(f"Artifact {artifact_id} is findable: {uri}")
            else:
                print(f"Artifact {artifact_id} is NOT findable: {uri}")

if __name__ == "__main__":
    json_file_path = 'artifacts.json'
    checker = FindabilityChecker(json_file_path)
    checker.check_findability()
```
